```python
import numpy as np
import pandas as pd
import difflib 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

```

```python
#read the csv file

data = pd.read_csv("ML/movies.csv")
```

```python
#first 5 rows

data.head()
```

```python
data.shape
```

```python
#select appropriate feature from data i.e feature selection

selected_features  = ['genres','keywords','tagline','cast','director']
print(selected_features)

```

```python
#replace null values by empty string

for f in selected_features:
  data[f]=data[f].fillna('')
```

```python
#combine selected features

combined_features = data['genres']+' '+data['keywords']+' '+data['tagline']+' '+data['cast']+' '+data['director']
print(combined_features)
```

```python
#transform textual data of combined feature into feature vector

vectoriser = TfidfVectorizer()
feature_vectors = vectoriser.fit_transform(combined_features)
print(feature_vectors)
```

```python
#get similarity score 

similarity = cosine_similarity(feature_vectors)

print(similarity)
```

```python
print(similarity.shape)
```

```python
#input movie name

movie_name = input('Enter your favorite movie name : ')
```

```python
#get movie name list

list_of_movie_name = data['title'].tolist()

print(list_of_movie_name)
```

```python
#match movie name given by user with movie name list 

match_movie = difflib.get_close_matches(movie_name,list_of_movie_name)

print(match_movie)
```

```python
close_match = match_movie[0]

print(close_match)
```

```python
#find index of movie name with title

i = data[data.title == close_match]['index'].values[0]

print(i)
```

```python
#get a list of similar movies

similarity_score = list(enumerate(similarity[i]))

print(similarity_score)

#first value of above print stmt represents index and second value is similarity score
#compared to movie name and given movie name
```

```python
#sort movies according to their similarity score

sort_similar_movies = sorted(similarity_score, key = lambda x : x[1] , reverse=True)

print(sort_similar_movies)
```

```python
#take name of similar movies based in index

print('Movies you may like :\n')

i=1
for movie in sort_similar_movies:
  index=movie[0]
  title_of_index = data[data.index==index]['title'].values[0]
  if i <=20:
    print(i,'',title_of_index)
    i+=1
  
```

```python
#MOVIE RECOMMONDATION SYSTEM

movie_name = input('Enter your favorite movie name : ')

list_of_movie_name = data['title'].tolist()

match_movie = difflib.get_close_matches(movie_name,list_of_movie_name)

close_match = match_movie[0]

i = data[data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[i]))

sort_similar_movies = sorted(similarity_score, key = lambda x : x[1] , reverse=True)

print('Movies you may like :\n')

i=1
for movie in sort_similar_movies:
  index=movie[0]
  title_of_index = data[data.index==index]['title'].values[0]
  if i <=20:
    print(i,'',title_of_index)
    i+=1
```
